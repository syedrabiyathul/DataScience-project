```python

!pip install yfinance pandas plotly
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
def get_real_time_data(ticker):
  data = yf.download(tickers=ticker,period='5d',interval='1m')
  return data
def calculate_moving_average(data, window):
  data['MA'] = data['Close'].rolling(window=window).mean()
  print(data)
  return data

def plot_real_time_data(data):
  fig=plt.figure()
  x=fig.add_subplot(111)
  x.plot(data['Open'],label='Open')
  x.plot(data['High'],label='High')
  x.plot(data['Low'],label='Low')
  x.plot(data['Close'],label='Close')

  x.xaxis.set_major_formatter(mdates.DateFormatter('%d'))


  plt.legend()
  plt.title('Real-Time Stock Data')
  plt.xlabel('interval')
  plt.xticks(rotation=45, ha='right')
  plt.ylabel('Price')
  plt.show()

data=get_real_time_data('AAPL')
data=calculate_moving_average(data,30)

plot_real_time_data(data)

```
