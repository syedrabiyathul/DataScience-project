```python
from googlesearch import search
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from textblob import TextBlob
import textstat
from readability.readability import Document


def get_search_results(query, num_results=5):
    results = []
    for result in search(query, num_results=num_results):
        results.append(result)
    return results


def get_website_content(url):
    try:
        response = requests.get(url, timeout=5)
        if response.status_code == 200:

            soup = BeautifulSoup(response.text, "html.parser")
            return soup.get_text()
        return None
    except requests.exceptions.RequestException:
        return None



def analyze_content(text):
    if not text:
        return {"readability_score": None, "sentiment": None}

    readability_score = textstat.flesch_reading_ease(text)
    sentiment = TextBlob(text).sentiment.polarity

    return {"readability_score": readability_score, "sentiment": sentiment}


def rank_websites(analysis_results):
    scores = {}
    for url, metrics in analysis_results.items():
        if metrics["readability_score"] is not None:
            score = (metrics["readability_score"] * weights["readability"]) + \
                    (metrics["sentiment"] * weights["sentiment"])
            scores[url] = score

    ranked = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    return ranked

weights = {"readability": 0.3, "sentiment": 0.7}


keyword = input('Enter a keyword')
websites = get_search_results(keyword, num_results=5)
print("Top Websites:", websites)

contents = {url: get_website_content(url) for url in websites}


analysis_results = {url: analyze_content(contents[url]) for url in websites}


ranked_sites = rank_websites(analysis_results)
print("Ranked Websites:", ranked_sites)
```
